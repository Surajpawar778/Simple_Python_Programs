# **1) Factorial of the number**

Prompt:- Write the python program to compute the factorial of the number , take the input from the user.

```python
def factorial(n):
    if not isinstance(n, int) or n < 0:
        return "Factorial is not defined for negative numbers or non-integers."
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Get input from the user
try:
    num = int(input("Enter a non-negative integer: "))
    print(f"The factorial of {num} is {factorial(num)}")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

# **2) Table of the given number**

Prompt:- Write the python program to display the multiplication table of the number,take the number from the user.  

```python
# Get input from the user
try:
    num = int(input("Enter a number to see its multiplication table: "))

    print(f"Multiplication Table for {num}:")
    # Iterate from 1 to 10
    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")

except ValueError:
    print("Invalid input. Please enter an integer.")
```

# **3) Even or Odd**

Prompt :- Write the python program to find if the given number is even or odd, take the number from the user.

```python
# Get input from the user
try:
    num = int(input("Enter an integer: "))

    # Check if the number is even or odd
    if (num % 2) == 0:
        print(f"{num} is an even number.")
    else:
        print(f"{num} is an odd number.")

except ValueError:
    print("Invalid input. Please enter an integer.")
```

# **4) Palindrome**

Prompt:- Write the program to find if the number is Palindrome or not , take the inputs from the user.

```python
# Get input from the user
try:
    num_str = input("Enter a number to check if it's a palindrome: ")

    # Check if the number string is a palindrome
    if num_str == num_str[::-1]:
        print(f"{num_str} is a palindrome.")
    else:
        print(f"{num_str} is not a palindrome.")

except ValueError:
    print("Invalid input. Please enter a valid number.")
```

# **5) Expert System**

Prompt:- Write the python program to make the expert system.

```python
# Simple Expert System for Disease Diagnosis

def expert_system():
    print("=== Medical Diagnosis Expert System ===")

    fever = input("Do you have fever? (yes/no): ").lower()
    cough = input("Do you have cough? (yes/no): ").lower()
    headache = input("Do you have headache? (yes/no): ").lower()
    sneezing = input("Do you have sneezing? (yes/no): ").lower()

    # Inference Engine (Rules)
    if fever == "yes" and cough == "yes" and headache == "yes":
        diagnosis = "You may have FLU."
    elif sneezing == "yes" and cough == "yes" and fever == "no":
        diagnosis = "You may have COMMON COLD."
    elif fever == "yes" and headache == "yes":
        diagnosis = "You may have VIRAL FEVER."
    else:
        diagnosis = "Condition unclear. Please consult a doctor."

    print("\nDiagnosis Result:")
    print(diagnosis)


# Run the expert system
expert_system()
```
